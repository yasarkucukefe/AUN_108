```python
# Yeni dosya oluşturmak için
dosya_output = open('yeni_csv_dosya.csv','w',encoding='utf-8',newline='')
```

```python
csv_writer = csv.writer(dosya_output, delimiter=',')
```

```python
csv_writer.writerow(['a','b','c'])
```

```python
for i in range(1,101):
    csv_writer.writerow([i,2*i,i**2])
```

```python
dosya_output.close()
```

```python
# Mevcut bir dosyaya ekleme yapmak için
dosya = open('yeni_csv_dosya.csv','a',encoding='utf-8',newline='') # a: append
csv_writer = csv.writer(dosya, delimiter=',')

for i in range(1,101):
    csv_writer.writerow([i,2*i,i**2])
    
dosya.close()
```

```python
# Bitcoin fiyatı
url = "https://api.coindesk.com/v1/bpi/currentprice.json"
resp = requests.get(url)
```

```python
resp.content
```

```python
bitcoin_data = json.loads(resp.content)
```

```python
print(bitcoin_data)
```

```python
bitcoin_data['bpi']['USD']['rate_float'] # Bitcoin fiyatı
```

```python
# Python Dictionary => JSON veri formatına dönüştürmek
dict = bitcoin_data['bpi']
json_format_bpi = json.dumps(dict) # dict => JSON'a dönüştür
print(json_format_bpi)
```

```python
print(type(json_format_bpi))
```
