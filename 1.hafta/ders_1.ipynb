```python
# KDV hesaplamas
fiyat = 100
kdv_oran = 0.18
toplam_fiyat = fiyat + fiyat * kdv_oran
print("Fiyat: "+str(fiyat)+". KDV dahil fiyat: "+str(toplam_fiyat))
print("Fiyat: {}. KDV dahil fiyat: {}".format(fiyat,toplam_fiyat))
print("Fiyat: {0}. KDV dahil fiyat: {1}".format(fiyat,toplam_fiyat))
print("Fiyat: {1}. KDV dahil fiyat: {0}".format(toplam_fiyat,fiyat))
print("Fiyat: {a}. KDV dahil fiyat: {b}".format(a=fiyat, b=toplam_fiyat))
print(f"Fiyat: {fiyat}. KDV dahil fiyat: {toplam_fiyat}")
```
