## Python Grafik Kütüphanesi
### Matplotlib https://matplotlib.org/stable/plot_types/index.html

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# y = x**2
x = np.linspace(0, 5, 11)
print(x)
```

```python
y = x ** 2
```

```python
print(y)
```

```python
for xd,yd in zip(x,y):
    print(f"{xd} => {yd}")
```

```python
# plot
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Grafik Başlık')
```

```python
plt.plot(x,y)
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Grafik Başlık (y=$x^{2}$)')
plt.show()
```

```python
plt.plot(x,y,linestyle='-.', color='r', marker='o')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
plt.plot()
```

## Matplotlib alternatif method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.9,0.9])

axes.plot(x, y, 'b')
axes.plot(y, x, 'g')
axes.set_xlabel('X etiketi')
axes.set_ylabel('Y etiketi')
axes.set_title('Başlık')
```

```python
f1 = x ** 2;
f2 = x ** 3;
```

```python
plt.plot(x, f1, label='x**2')
plt.plot(x, f2, label='x***2')
plt.legend()
```

```python
# scatter
plt.scatter(x,f1)
```

## Seaborn https://seaborn.pydata.org/tutorial.html

```python
import seaborn as sns
%matplotlib inline
```

```python
bahsisler = sns.load_dataset('tips')
```

```python
type(bahsisler)
```

```python
bahsisler.head()
```

```python
sns.displot(bahsisler['tip'],bins=50)
```

```python
sns.jointplot(x='total_bill', y='tip', data=bahsisler, kind='scatter')
```

```python
sns.pairplot(bahsisler)
```

```python
sns.pairplot(bahsisler, hue='smoker', palette='coolwarm')
```

```python
sns.barplot(x='sex', y='total_bill', data=bahsisler)
```

```python
sns.countplot(x='smoker', data=bahsisler)
```

```python
sns.countplot(x='day', data=bahsisler)
```

```python
sns.boxplot(x='day', y='total_bill', data=bahsisler)
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=bahsisler)
```
